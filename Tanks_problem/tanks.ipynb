# Two Tanks Draining Problem

## Which one epmties first?

<img src="images/Tanks.png" width="600">

## Geometry of the System

Both tanks:

- Have a hole at the bottom.
- The hole is circular.
- The hole radius is smaller than the radius of the tank base.
- The hole area is the same for both tanks.

Let:

$$
A = \pi r_h^2
$$

where:

- $r_h$ = radius of the hole  
- $A$ = hole area (same for both tanks)

---

# Physical Assumptions

- Incompressible fluid
- Ideal flow (no viscosity)
- Atmospheric pressure at surface and at the hole
- The hole is located at the bottom
- The hole size is small compared to the tank section

---

# Torricelli’s Law

The exit velocity is:

$$
v = \sqrt{2 g h}
$$

where:

- $h(t)$ = water height
- $g$ = gravitational acceleration

---

# Flow Rate

The volumetric flow rate is:

$$
Q = A \sqrt{2 g h}
$$

Since the hole is identical in both tanks,  
the hydraulic discharge law is the same.

---

# Volume Balance

Let $V(h)$ be the water volume.

The volume decreases:

$$
\frac{dV}{dt} = - A \sqrt{2 g h}
$$

---

# Geometric Relation

For each tank:

$$
\frac{dV}{dt} = S(h) \frac{dh}{dt}
$$

where:

- $S(h)$ = cross-sectional area at height $h$

---

# Differential Equation

$$
S(h) \frac{dh}{dt} = - A \sqrt{2 g h}
$$

or

$$
\frac{dh}{dt} =
- \frac{A}{S(h)} \sqrt{2 g h}
$$

---

# Key Point

Since the hole area $A$ is the same for both tanks,  
the only difference in the draining behavior comes from:

$$
S(h)
$$

The geometry determines how fast the height decreases over time.

---

---

# Cross-sectional Area for a Conical Frustum (Tank Height Reference)

Each tank is a **conical frustum** (truncated cone).

Let:

- $H_T$ = total geometric height of the tank  
- $h_0$ = initial water height  
- $h(t)$ = water height measured from the bottom (hole) upward  
- $r_b$ = radius at the bottom section ($h=0$)  
- $r_t$ = radius at the top section of the tank ($h=H_T$)  

We assume:

$$
0 \le h(t) \le h_0 \le H_T
$$

and the initial condition:

$$
h(0)=h_0
$$

---

## Radius as a function of height

The radius varies linearly along the full tank height $H_T$:

$$
r(h)=r_b+\frac{r_t-r_b}{H_T}\,h
$$

Check:

- $r(0)=r_b$
- $r(H_T)=r_t$

---

## Cross-sectional area

The cross-sectional area at height $h$ is:

$$
S(h)=\pi\,r(h)^2
$$

Therefore:

$$
S(h)=\pi\left(r_b+\frac{r_t-r_b}{H_T}\,h\right)^2
$$

---
---
---

# Tank Design and Determination of the Initial Water Height $h_0$

We want the system to contain:

$$
V_0 = 1000 \text{ liters}
$$

To make the system feasible, we assume that the tank capacity is larger than the water volume.

We choose:

$$
V_{tank} = 1.5 \, V_0
$$

Therefore:

$$
V_{tank} = 1500 \text{ liters}
$$

---

## Unit Conversion

$$
1 \text{ m}^3 = 1000 \text{ L}
$$

So:

$$
V_0 = 1 \text{ m}^3
$$

$$
V_{tank} = 1.5 \text{ m}^3
$$

---

# Step 1 — Determination of the Tank Height $H_T$

To define a physically consistent tank geometry, we proceed as follows.

For modeling convenience, we **fix the bottom and top radii** of the conical frustum:

- bottom radius $r_b$
- top radius $r_t$

These two parameters define the *shape* (opening angle) of the tank.

However, in order for the tank to be physically realizable and able to contain the prescribed amount of water, the **total height $H_T$** must be determined such that the tank volume equals the desired volume $V_{tank}$.

In other words:

> The geometry (radii) is fixed first,  
> then the height is computed to ensure volumetric consistency.

---

## Volume of a Conical Frustum

The volume of a conical frustum is:

$$
V_{tank} = \frac{\pi H_T}{3}
\left(r_b^2 + r_b r_t + r_t^2\right)
$$

---

## Solving for the Required Height

Since $V_{tank}$, $r_b$, and $r_t$ are specified, we solve the volume equation for $H_T$:

$$
H_T =
\frac{3 V_{tank}}
{\pi \left(r_b^2 + r_b r_t + r_t^2\right)}
$$

This guarantees that the chosen geometry can physically contain the target volume.

---

## Physical Interpretation

- Larger radii imply a smaller required height for the same volume.
- Smaller radii require a taller tank to store the same amount of water.
- The height is therefore **not arbitrary**, but dictated by volumetric consistency.

---

## Numerical Example

Choose:

$$
r_b = 0.2 \text{ m}
$$

$$
r_t = 0.5 \text{ m}
$$

Compute:

$$
r_b^2 = 0.04
$$

$$
r_b r_t = 0.10
$$

$$
r_t^2 = 0.25
$$

Sum:

$$
0.39
$$

Substitute:

$$
H_T =
\frac{3 \cdot 1.5}
{\pi \cdot 0.39}
$$

Numerically:

$$
H_T \approx 3.67 \text{ m}
$$

---

# Step 2 — Determine the Initial Water Height $h_0$

Now we find $h_0$ such that the water volume is:

$$
V(h_0) = V_0 = 1 \text{ m}^3
$$

The radius varies linearly:

$$
r(h)=r_b+\frac{r_t-r_b}{H_T}\,h
$$

The free-surface radius is:

$$
r_0=r(h_0)
$$

The water volume up to height $h_0$ is:

$$
V(h_0)=
\frac{\pi h_0}{3}
\left(r_b^2+r_b r_0+r_0^2\right)
$$

We solve:

$$
V(h_0)=1
$$

This equation is cubic in $h_0$, so we compute $h_0$ numerically (for example using the bisection method).

---

# Tank Y Case

For Tank Y we swap the radii:

$$
r_b = 0.5 \text{ m}, \quad r_t = 0.2 \text{ m}
$$

The tank height $H_T$ remains the same (because the total capacity is the same).

However, the function $V(h)$ changes shape, so the resulting $h_0$ will generally be different.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Water volume in  liters contained in the tank
initial_water_liters = 1000

#The tank must contain the whole volume of water
tank_capacity = initial_water_liters*1.5

low_radius = 0.2  # [m]
high_radius = 0.5 # [m]

def calc_tanks_height(tank_capacity,r1,r2):
    V = tank_capacity/1000.0 # convert liters to m^3
    return 3*V/(np.pi*(r1**2+r2**2+r1*r2))

def calc_r_h(h,rb,rt,tank_height):
    return rb + (rt-rb)*(h/tank_height)

# calculate the height of the tank that contains 1.5 times the amount of water
H_T = calc_tanks_height(tank_capacity,low_radius,high_radius)


#Tank X geometry definition
r_b_tank_x = low_radius
r_t_tank_x = high_radius

#Tank Y geometry definition
r_b_tank_y = high_radius
r_t_tank_y = low_radius

# hole radius (same for both tanks)
r_hole = 0.05  # [m]

print(f'Water volume in each tank: {initial_water_liters} [l]')
print(f'Tanks height: {H_T:.2f} [m]')

print(f'Tank X bottom section diameter: {2*r_b_tank_x} [m]')
print(f'Tank X top section diameter: {2*r_t_tank_x} [m]')

print(f'Tank Y bottom section diameter: {2*r_b_tank_y} [m]')
print(f'Tank Y top section diameter: {2*r_t_tank_y} [m]')
```

# Finding $h_0$ with the Bisection Method

We want to find the initial water height $h_0$ such that the water volume inside the tank is exactly the target volume $V_0$.

---

## 1) Geometry and Volume Function

The tank radius changes linearly with height:

$$
r(h)=r_b+\frac{r_t-r_b}{H_T}\,h
$$

The cross-sectional area is:

$$
S(h)=\pi r(h)^2
$$

The volume of water up to height $h$ (a frustum from $0$ to $h$) is:

$$
V(h)=\frac{\pi h}{3}\left(r_b^2+r_b r(h)+r(h)^2\right)
$$

---

## 2) The Root-Finding Problem

We define a function:

$$
f(h)=V(h)-V_0
$$

We want:

$$
f(h_0)=0
$$

So the problem becomes: **find the root of $f(h)$** in the interval $[0, H_T]$.

---

## 3) Bracketing the Root

First we check the values at the interval boundaries:

$$
f(0)=V(0)-V_0=0-V_0=-V_0<0
$$

and:

$$
f(H_T)=V(H_T)-V_0
$$

We must have:

$$
f(H_T)\ge 0
$$

meaning the tank capacity is at least $V_0$.

So the root is inside $[0, H_T]$ if:

$$
f(0)\cdot f(H_T)\le 0
$$

---

## 4) Bisection Algorithm (Step-by-Step)

We start with:

$$
h_L = 0,\quad h_U = H_T
$$

At each iteration:

1) Compute the midpoint:

$$
h_M=\frac{h_L+h_U}{2}
$$

2) Evaluate:

$$
f(h_M)=V(h_M)-V_0
$$

3) Update the interval:

- If $f(h_M)=0$, we found the solution: $h_0=h_M$.
- If $f(h_L)\,f(h_M)<0$, the root is in $[h_L,h_M]$, so set:

$$
h_U \leftarrow h_M
$$

- If $f(h_L)\,f(h_M)>0$, the root is in $[h_M,h_U]$, so set:

$$
h_L \leftarrow h_M
$$

4) Repeat until the interval is small enough:

$$
|h_U-h_L| < \varepsilon
$$

Then we take:

$$
h_0 \approx \frac{h_L+h_U}{2}
$$

---

## 5) Why Bisection Works Here

- $V(h)$ increases with $h$ (more height means more volume)
- therefore $f(h)=V(h)-V_0$ is increasing
- if $f(0)<0$ and $f(H_T)>0$, there is exactly one solution in $[0,H_T]$

So bisection is **guaranteed** to converge.

---

## 6) Final Output

The result of the method is the height:

$$
h_0
$$

such that:

$$
V(h_0)=V_0
$$

```python
# total volume of the water in the tank at height h
def calc_V_h(h,rb,rt,tank_height):
    # params:
    # h:  evaluation point
    # rb: bottom radius of the tank
    # rt: top radius of the tank
    # tank_height: height of the tank

    # returns:
    # volume of the water whose surface is at height h

    #calculate the radius of the section of the tank corresponding to the height h
    r_h = calc_r_h(h,rb,rt,tank_height)
    # return the volume
    return ((np.pi*h)/3)*(rb**2+rb*r_h+r_h**2)

# find the height of the water h0 corresponding to the desired volume V0
# bisection method implemented
def calc_h_at_V0(V0, rb, rt, tank_height, max_iter=200, xtol=1e-8, ytol=1e-12):
    # params:
    # V0: water volume inside the tank [m^3]
    # rb: bottom radius of the tank
    # rt: top radius of the tank
    # tank_height: height of the tank
    # max_iter: maximum number of iterations if the bisection method
    # xtol : absolute tolerance on the domain variable (h). The algorithm stops if the current interval length becomes smaller than xtol.
    # ytol : absolute tolerance on the function value. The algorithm stops if |f(h)| becomes smaller than ytol, meaning the residual volume error is sufficiently small.

    # returns:
    # height of the water surface if its volume is V0

    low_lim, high_lim = 0.0, tank_height

    f_low  = calc_V_h(low_lim,  rb, rt, tank_height) - V0
    f_high = calc_V_h(high_lim, rb, rt, tank_height) - V0

    #check if f_low is a root
    if abs(f_low) < ytol:
        return low_lim
    
    #check if f_high is a root
    if abs(f_high) < ytol:
        return high_lim
    
    #check the zero-crossing feasibility
    if f_low * f_high > 0:
        raise ValueError("Root out of range: check V0 or tank capacity")

    for _ in range(max_iter):
        
        #function evaluation at the mid point
        mid = 0.5*(low_lim + high_lim)
        f_mid = calc_V_h(mid, rb, rt, tank_height) - V0

        #checl if the mid point is a root or if it's very close to actual root
        if abs(f_mid) < ytol or (high_lim - low_lim) < xtol:
            return mid

        #update the limits if the zero condition has not been verified
        if f_low * f_mid < 0:
            high_lim = mid
            f_high = f_mid
        else:
            low_lim = mid
            f_low = f_mid

    return 0.5*(low_lim + high_lim)

# find the initial height h0 of the water surface corresponding to the desired volume of V0 in Tank X
h0_tank_x = calc_h_at_V0(initial_water_liters/1000,r_b_tank_x,r_t_tank_x,H_T)
# find the corresponding section radius at h0 in Tank X
r0_tank_x = calc_r_h(h0_tank_x,r_b_tank_x,r_t_tank_x,H_T)

print(f'Initial height of water surface in tank X: {h0_tank_x:.2f} [m]')
print(f'r(h0) tank X: {r0_tank_x:.2f} [m]')

# find the initial height h0 of the water surface corresponding to the desired volume of V0 in Tank Y
h0_tank_y = calc_h_at_V0(initial_water_liters/1000,r_b_tank_y,r_t_tank_y,H_T)
# find the corresponding section radius at h0 in Tank Y
r0_tank_y = calc_r_h(h0_tank_y,r_b_tank_y,r_t_tank_y,H_T)

print(f'Initial height of water surface in tank Y: {h0_tank_y:.2f} [m]')
print(f'r(h0) tank Y: {r0_tank_y:.2f} [m]')
```

```python
# draw the tank in the figure
def draw_tank(ax, cx, rb, rt, tank_height, h0, r_hole,label,n=400, water_color="tab:blue"):
    #params:
    # ax : plot figure reference
    # cx : offset in x coordinate of the simmetry axis of the tank
    # rb : bottom radius of the tank
    # rt : top radius of the tank
    # tank_height : height of the tank
    # h0: height of the water surface
    # r_hole: radius of the hole in the bottom surface
    # label: label of the tank
    # n : number of discretized points on the y-axis
    # water_color : desired color of the water

    # array of discretized h (vertical coordinates)
    h = np.linspace(0, tank_height, n)
    # array of radii associated to the values of h
    r = calc_r_h(h, rb, rt, tank_height)

    # x-coordinates of the left line representing the wall of the tank (shifted by cx)
    xL = cx - r
    # x-coordinates of the right line representing the wall of the tank (shifted by cx)
    xR = cx + r

    # plot tank walls
    ax.plot(xL, h, color="k", lw=2)
    ax.plot(xR, h, color="k", lw=2)

    # bottom: open in the middle (hole)
    # left bottom segment: from (cx-rb) to (cx-r_hole)
    ax.plot([cx - rb, cx - r_hole], [0, 0], color="k", lw=2)
    # right bottom segment: from (cx+r_hole) to (cx+rb)
    ax.plot([cx + r_hole, cx + rb], [0, 0], color="k", lw=2)

    # a short vertical tick to show the opening
    ax.plot([cx - r_hole, cx - r_hole], [0, -0.03*tank_height], color="k", lw=2)
    ax.plot([cx + r_hole, cx + r_hole], [0, -0.03*tank_height], color="k", lw=2)

    # water fill (0 -> h0)
    hw = np.linspace(0, h0, n)
    rw = calc_r_h(hw, rb, rt, tank_height)
    xLw = cx - rw
    xRw = cx + rw
    ax.fill_betweenx(hw, xLw, xRw, color=water_color, alpha=0.35)

    # water surface line
    r0 = calc_r_h(h0, rb, rt, tank_height)
    ax.plot([cx - r0, cx + r0], [h0, h0], color=water_color, lw=2)

    # label
    ax.text(cx, tank_height + 0.05*tank_height, label, ha="center", va="bottom")


# plot initialization
fig, ax = plt.subplots(figsize=(10, 6))

# x-coordinate shift for the tank X
cx_tank_x = -2*high_radius
cx_tank_y = +2*high_radius

draw_tank(ax, cx_tank_x, r_b_tank_x, r_t_tank_x, H_T, h0_tank_x, r_hole,"Tank X")
draw_tank(ax, cx_tank_y, r_b_tank_y, r_t_tank_y, H_T, h0_tank_y, r_hole,"Tank Y")

ax.set_xlabel("x [m]")
ax.set_ylabel("Water surface height [m]")
ax.set_title(f"Two tanks with same amount of water ({initial_water_liters:.1f} l)")

# update x limits based on actual tank centers
x_min = min(cx_tank_x - max(r_b_tank_x, r_t_tank_x), cx_tank_y - max(r_b_tank_y, r_t_tank_y)) - 0.2
x_max = max(cx_tank_x + max(r_b_tank_x, r_t_tank_x), cx_tank_y + max(r_b_tank_y, r_t_tank_y)) + 0.2
ax.set_xlim(x_min, x_max)

ax.set_ylim(-0.15*H_T, 1.15*H_T)

# hide x-axis info and grid
ax.get_xaxis().set_visible(False)
ax.grid(False)

ax.set_aspect("equal", adjustable="box")

plt.show()
```

# Mathematical Interlude: Euler Method for a First-Order ODE

We consider a general first-order ordinary differential equation (ODE):

$$
\dot y = \frac{dy}{dt} = f(t,y)
$$

with initial condition:

$$
y(t_0) = y_0
$$

---

## Time Discretization

We introduce a small time step:

$$
\epsilon > 0
$$

and define discrete time instants:

$$
t_n = t_0 + n \epsilon
$$

where:

$$
n = 0,1,2,\dots
$$

We denote:

$$
y_n \approx y(t_n)
$$

---

## Idea of the Euler Method

From the definition of derivative:

$$
\frac{dy}{dt} \approx \frac{y_{n+1} - y_n}{\epsilon}
$$

Substituting into the ODE:

$$
\frac{y_{n+1} - y_n}{\epsilon} = f(t_n, y_n)
$$

---

## Euler Update Formula

Solving for $y_{n+1}$:

$$
y_{n+1} = y_n + \epsilon\, f(t_n, y_n)
$$

This is the **explicit Euler method**.

---

## Interpretation

- We start from $(t_n, y_n)$.
- We compute the slope $f(t_n, y_n)$.
- We move forward by a small time step $\epsilon$.
- We obtain the new approximation $y_{n+1}$.

---

## Application to the Tank Problem

For the draining tank:

$$
\frac{dh}{dt} =
- \frac{A}{S(h)} \sqrt{2 g h}
$$

The Euler scheme becomes:

$$
h_{n+1}
=
h_n
-
\epsilon
\frac{A}{S(h_n)} \sqrt{2 g h_n}
$$

where $\epsilon$ is the time step.

Thus

$$
f(t_n,h_n) = - \frac{A}{S(h_n)} \sqrt{2 g h_n} = -\frac{r_{hole}^2}{r(h_n)^2} \sqrt{2 g h_n}
$$

```python
# generic first order ODE implementing the explicit Euler method
def first_order_ODE_solver(f,t0,y0,epsilon,max_time,*f_args):
    # parameters:
    # f:  f(t,y) pointer
    # t0: initial time t0
    # y0: initial condition y(t0)
    # epsilon: discretization step
    # max_time: final time tf where the solver stops
    # f_args: additional arguments for the functuon f(t,y)

    # returns:
    # time array
    # y solution array

    t = [t0]
    y = [y0]
    while t[-1]<max_time:
        y.append(y[-1]+epsilon*f(t[-1],y[-1],*f_args))
        t.append(t[-1]+epsilon)
    return np.array(t), np.array(y)

# definition of the function f(t,y) for the tank problem
def tank_f(t,h,r_hole,g,rb,rt,tank_height):
    # parameters:
    # t: time for evaluating f(t,y)
    # y: y value for evaluating f(t,y)
    # r_hole: hole radius of the tank
    # g: gravitational acceleration
    # rb: bottom radius of the tank
    # rt: top radius of the tank
    # tank_height: height of the tank

    # returns:
    # f(t,y)

    # if height is less than zero, then the tank is empty.
    if h<=0:
        return 0.0
    r_of_h = calc_r_h(h,rb,rt,tank_height)
    return -((r_hole/r_of_h)**2)*np.sqrt(2*g*h)
```

```python
epsilon = 0.01 # time discretization step - 10 ms
observation_time = 60 # 60 s

# get t,h (solution) for the tank X
tank_x_t, tank_x_h =  first_order_ODE_solver(
    tank_f,0,h0_tank_x,epsilon,observation_time,r_hole,9.81,r_b_tank_x,r_t_tank_x,H_T
)

# get t,h (solution) for the tank Y
tank_y_t, tank_y_h =  first_order_ODE_solver(
    tank_f,0,h0_tank_y,epsilon,observation_time,r_hole,9.81,r_b_tank_y,r_t_tank_y,H_T
)
```

```python
plt.xlabel("time [s]")
plt.ylabel("Water surface height [m]")
plt.plot(tank_x_t,tank_x_h,label=r"$h_X(t)$")
plt.plot(tank_y_t,tank_y_h,label=r"$h_Y(t)$")
plt.legend()
plt.show()
```

```python
#Draw static tank geometry and create dynamic water artists
#Returns a dictionary with everything needed for animation updates
def init_tank_artists(ax, cx, rb, rt, tank_height, r_hole, label, n=400, water_color="tab:blue"):
    # params
    # ax : plot figure reference
    # cx : offset in x coordinate of the simmetry axis of the tank
    # rb : bottom radius of the tank
    # rt : top radius of the tank
    # tank_height : height of the tank
    # h0: height of the water surface
    # r_hole: radius of the hole in the bottom surface
    # label: label of the tank
    # n : number of discretized points on the y-axis
    # water_color : desired color of the water

    # returns:
    # dictionary of the grafical definitions for the tank

    # array of discretized h (vertical coordinates)
    h = np.linspace(0, tank_height, n)
    # array of radii associated to the values of h
    r = calc_r_h(h, rb, rt, tank_height)

    # x-coordinates of the left line representing the wall of the tank (shifted by cx)
    xL = cx - r
    # x-coordinates of the right line representing the wall of the tank (shifted by cx)
    xR = cx + r

    # ---- STATIC PARTS ----
    # walls
    ax.plot(xL, h, color="k", lw=2)
    ax.plot(xR, h, color="k", lw=2)

    # bottom with hole gap
    ax.plot([cx - rb, cx - r_hole], [0, 0], color="k", lw=2)
    ax.plot([cx + r_hole, cx + rb], [0, 0], color="k", lw=2)

    # hole markers
    ax.plot([cx - r_hole, cx - r_hole], [0, -0.03*tank_height], color="k", lw=2)
    ax.plot([cx + r_hole, cx + r_hole], [0, -0.03*tank_height], color="k", lw=2)

    # label
    ax.text(cx, tank_height + 0.05*tank_height, label, ha="center", va="bottom")

    # ---- DYNAMIC WATER OBJECTS (initially empty) ----
    
    # Start with h = 0 (bottom of the tank)

    #water surface height (y coordinate)
    hw = np.array([0.0]) # bottom of the tank
    # corresponding section radius at the water surface height
    rw = calc_r_h(hw, rb, rt, tank_height) #returns rb

    #left x coordinate of the water at the surface
    xLw = cx - rw
    #right x coordiante of the water at the surface
    xRw = cx + rw

    # fills the area of a polygon. Here are inserted the two points of the bottom of the tank
    # when updating, the polygon fill will be done using the real water height considered in the update phase
    # in this specific case, area = 0, no area to fill
    water_poly = ax.fill(
        np.concatenate([xLw, xRw[::-1]]), #polygon vertices x coordinates
        np.concatenate([hw, hw[::-1]]),   #polygon vertices y coordinates
        color=water_color,
        alpha=0.35
    )[0]

    # water line, initiliazed at the bottom of the tank (it will be updated later)
    water_line, = ax.plot(
        [xLw[0], xRw[0]],
        [0, 0],
        color=water_color,
        lw=2,
        alpha=0.35
    )

    return {
        "cx": cx,
        "rb": rb,
        "rt": rt,
        "tank_height": tank_height,
        "n": n,
        "water_poly": water_poly,
        "water_line": water_line,
        "r_hole": r_hole
    }
```

```python
#Update water polygon and surface line for given height of the water surface
def update_water(artists, h_level):
    #params:
    # artists: artirts reference to update
    # h_level: actual water level for which the update must be done

    #returns:
    # updated artists

    #retrieves the artists data
    cx = artists["cx"]
    rb = artists["rb"]
    rt = artists["rt"]
    tank_height = artists["tank_height"]
    n = artists["n"]
    r_hole = artists["r_hole"]

    #assure h_level is between 0 and tank_height
    h_level = float(max(0.0, min(h_level, tank_height)))

    #retrieves the coordinates of the points of the polygon for water filling
    #y coordinates
    hw = np.linspace(0, h_level, n)
    #x coordinates
    rw = calc_r_h(hw, rb, rt, tank_height)

    #shifting the x coordinates for the considered tank
    xLw = cx - rw
    xRw = cx + rw

    # create the polygon points
    X = np.concatenate([xLw, xRw[::-1]])
    Y = np.concatenate([hw, hw[::-1]])
    #update the water polygon to fill
    artists["water_poly"].set_xy(np.column_stack([X, Y]))

    #update the water surface line
    r0 = calc_r_h(h_level, rb, rt, tank_height)

    #if h_level is greater than 0 then draw it, otherwise remove the line
    if h_level>0.0:
        artists["water_line"].set_data(
            [cx - r0, cx + r0],
            [h_level, h_level]
        )
    else:
        artists["water_line"].set_visible(False)

    return artists["water_poly"], artists["water_line"]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

cx_X = -2*high_radius
cx_Y = +2*high_radius

# static + dynamic artists
artists_X = init_tank_artists(ax, cx_X, r_b_tank_x, r_t_tank_x, H_T, r_hole, "Tank X")
artists_Y = init_tank_artists(ax, cx_Y, r_b_tank_y, r_t_tank_y, H_T, r_hole, "Tank Y")

ax.set_title("Draining tanks\nsame hole at the bottom\nsame initial water volume)")
ax.set_aspect("equal", adjustable="box")
ax.grid(False)
ax.get_xaxis().set_visible(False)

# good limits
max_r = max(r_b_tank_x, r_t_tank_x, r_b_tank_y, r_t_tank_y)
x_min = min(cx_X - max_r, cx_Y - max_r) - 0.3
x_max = max(cx_X + max_r, cx_Y + max_r) + 0.3
ax.set_xlim(x_min, x_max)
ax.set_ylim(-0.15*H_T, 1.15*H_T)

# remove y ticks too for a cleaner representation
ax.set_yticks([])

# time text
time_text = ax.text(0.42, 0.95, "", transform=ax.transAxes)

# animate until both are empty
T_end = max(tank_x_t[idx_X], tank_y_t[idx_Y])

# set animation frames per second
fps = 60
t_anim = np.linspace(0, T_end, int(T_end*fps)+1)

# interpolate heights to the common timeline
hX_anim = np.interp(t_anim, tank_x_t, tank_x_h)
hY_anim = np.interp(t_anim, tank_y_t, tank_y_h)

def init():
    # initialize with first frame values
    update_water(artists_X, hX_anim[0])
    update_water(artists_Y, hY_anim[0])
    time_text.set_text(f"t = {t_anim[0]:.1f} s")
    return (*update_water(artists_X, hX_anim[0]), *update_water(artists_Y, hY_anim[0]), time_text)

def update(i):
    update_water(artists_X, hX_anim[i])
    update_water(artists_Y, hY_anim[i])
    time_text.set_text(f"t = {t_anim[i]:.1f} s")
    return (*update_water(artists_X, hX_anim[i]), *update_water(artists_Y, hY_anim[i]), time_text)

anim = FuncAnimation(fig, update, frames=len(t_anim), init_func=init, blit=True, interval=1000/fps)

plt.show()
```
